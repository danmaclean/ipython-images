```python
def max_proj(img_list):
    return np.maximum.reduce(img_list)

a = max_proj(flex)

from skimage import exposure
b = exposure.rescale_intensity(a, in_range='uint16')
io.imshow(b)

```

```python
from skimage.restoration import denoise_tv_chambolle, denoise_bilateral
smooth = denoise_bilateral(b, sigma_range=0.7, sigma_spatial=15)
io.imshow(smooth)
```

```python
from skimage import data, filter, color
from skimage.transform import hough_circle
from skimage.feature import peak_local_max
from skimage.draw import circle_perimeter
from skimage.util import img_as_ubyte
from skimage import data, filter, color
from skimage.transform import hough_ellipse
from skimage.draw import ellipse_perimeter
```

```python
edges = filter.canny(small, sigma=1,
                     low_threshold=0.002, high_threshold=0.005)
io.imshow(edges)
```
